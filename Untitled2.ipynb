```python
df= pd.read_csv ("opencell-protein-interactions.csv")
df.head(10)
```

```python
cdf = df[['pval','enrichment','interaction_stoichiometry','abundance_stoichiometry']]
cdf.head(100)
```

```python
plt.scatter(cdf.enrichment , cdf.pval , color= 'blue')
plt.xlabel("enrichment")
plt.ylabel("pval")
plt.show()
```

```python
msk = np.random.rand(len(df)) < 0.8
train = cdf [msk]
test = cdf [~msk]
```

```python
from sklearn import linear_model
import numpy as np 
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['enrichment']])
train_y = np.asanyarray(train[['pval']])
regr. fit (train_x , train_y)
print ('coefficients:', regr.coef_)
print('Intercept:' , regr.intercept_)
```

```python
plt.scatter(train.enrichment ,train.pval ,color = 'blue')
plt.plot(train_x ,regr.coef_[0][0]*train_x + regr.intercept_[0],'-r')
plt.xlabel("enrichmnet")
plt.ylabel("pval")
```

```python
from sklearn.metrics import r2_score

test_x = np.asanyarray (test[['enrichment']])
test_y =np.asanyarray(test[['pval']])
test_y_ =regr.predict(test_x)

print("Mean absolute error : %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("Resodial sum of sqares (MSE) : %.2f" % np.mean((test_y_ - test_y)**2))
print("R2-xsore:%2f" % r2_score (test_y , test_y_))
```
